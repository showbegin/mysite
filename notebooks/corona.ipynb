## Importing all required libraries and defining the classes for comfortable operating the data.

```python
from selenium import webdriver 
from selenium.webdriver.common.by import By 
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC 
from selenium.common.exceptions import TimeoutException
import time
import pytz
from datetime import datetime
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
import json
import math

class Time_point(object):
    '''Contains one time point data with Covid-19 statistics. An object for accumulation to the Covid19 class.
    
    Methods:
    show_time(): show timepoint
    show_area(area): show number of cases in the called area for the current timepoint.
    '''
    def __init__(self, time,stats):
        self.time = time
        self.stats = stats
        self.total = sum(self.stats.values())
    def show_time(self):
        return datetime.strftime(self.time, "%d/%m/%Y, %H:%M:%S")
    def show_area(self, area):
        try:
            return self.stats[area]
        except:
            print('Нет такого региона')
            
class Covid19(object):
    '''Accumulates the scrapped Covid-19 data from the dictionary.
    
    Methods:
    show(): prints start timepoint and final timepoint.
    show_all(): prints each row of statistics.
    save(): saves the accumulated data into the file.
    load(): loads the new data to the current object from the dictionary.
    make_dict(): creates the dictionary from the object data
    make_df(): make pandas.DataFrame from the object data
    '''
    def __init__(self, mode='infected', dic=None, path=None):
        '''Object constuctor.
        
        Args:
          mode (string): needs for methods that print information about data. If "infected" is set (default) - \
            this information will be print, else - information about "recovered" is shown.
          dic (dictionary): the dictionary from which the data are collected.
          path (string): The file path. If set instead of dic, the data from the file will be collected to the object.
        '''
        self.mode=mode
        try:
            if path:
                with open(self.mode+".json") as f:
                    dic=json.load(f)
                    dic2={}
                    for strkey, value in dic.items():
                        key=datetime.strptime(strkey, '%d-%m-%Y %H:%M').replace(tzinfo=pytz.utc)
                        dic2[key]=value
                    print("data is loaded from ",path)
                    dic=dic2
            self.points=[Time_point(time,stats) for time, stats in sorted(dic.items())]
            self.show()
        except:
            print("dic or path parameter must be set")
    def show(self):
        message = "Статистика с {} по {}".format(
            self.points[0].show_time(),
            self.points[-1].show_time()
        )
        print(message)
    def show_all(self):
        message = "Статистика: \n"
        for point in self.points:
            if self.mode=="infected":
                message = message+point.show_time()+' - всего {} инфицированных \n'.format(point.total)
            if self.mode=="recovered":
                message = message+point.show_time()+' - всего {} выздоровевших \n'.format(point.total)
        print(message)
    def save(self):
        jsoned={}
        for point in self.points:
            key=datetime.strftime(point.time, '%d-%m-%Y %H:%M')
            jsoned[key]=point.stats
        with open(self.mode+".json", mode="w") as f:
            json.dump(jsoned, f)
        print("data is saved into the "+self.mode+".json")
    def load(self,newdic):
        dic={}
        for point in self.points:
            dic[point.time]=point.stats
        dic={**dic,**newdic}
        self.points=[Time_point(time,stats) for time, stats in sorted(dic.items())]
        self.show()
    def make_dict(self):
        dic={}
        for point in self.points:
            dic[point.time]=point.stats
        return dic
    def make_df(self, daily=True):
        '''makes pandas.DataFrame from the internal content of object
        
        Args:
          daily (boolean): if True (default) convolves time dimention to the days. Aggregates the values inside the \
            group by the max.
        
        Returns:
          pandas.DataFrame object
        '''
        dic=self.make_dict()
        df=pd.DataFrame(dic)
        df=df.sort_index(axis=1)
        df=df.fillna(0)
        df=df.astype('int16')
        df.index.name="Регион"
        df.columns.name="Дата"

        if daily==True:
            dl=df.copy().T
            dl.index=df.T.reset_index()["Дата"].apply(lambda x: x.date())
            dl=dl.groupby("Дата").agg(max).T
            df=dl
        return df

def parse_data(str):
    '''parses data from the news text
    
    Args: 
      str (string): text content of news
      
    Returns:
      dictionary containing areas as keys and the parsed number as values'''
    patterns=(
            (re.compile(r'\b(Нур\-?\w+).+?(\d+)\b'),'Нур-Султан'),
            (re.compile(r'\b(Алматы).+?(\d+)\b'),'Алматы'),
            (re.compile(r'\b(Шымк\w+).+?(\d+)\b'),'Шымкент'),
            (re.compile(r'\b(Акмол\w+).+?(\d+)\b'),'Акмолинская область'),
            (re.compile(r'\b(Актю\w+).+?(\d+)\b'),'Актюбинская область'),
            (re.compile(r'\b(Алмати\w+).+?(\d+)\b'),'Алматинская область'),
            (re.compile(r'\b(Атыраус\w+).+?(\d+)\b'),'Атырауская область'),
            (re.compile(r'\b(Восточн\w+).+?(\d+)\b'),'Восточно-Казахстанская область'),
            (re.compile(r'\b(Жамбылс\w+).+?(\d+)\b'),'Жамбылская область'),
            (re.compile(r'\b(Западн\w+).+?(\d+)\b'),'Западно-Казахстанская область'),
            (re.compile(r'\b(Караганд\w+).+?(\d+)\b'),'Карагандинская область'),
            (re.compile(r'\b(Костан\w+).+?(\d+)\b'),'Костанайская область'),
            (re.compile(r'\b(Кызыло\w+).+?(\d+)\b'),'Кызылординская область'),
            (re.compile(r'\b(Мангист\w+).+?(\d+)\b'),'Мангистауская область'),
            (re.compile(r'\b(Павлод\w+).+?(\d+)\b'),'Павлодарская область'),
            (re.compile(r'\b(Северо\w*).+?(\d+)\b'),'Северо-Казахстанская область'),
            (re.compile(r'\b(Туркест\w+).+?(\d+)\b'),'Туркестанская область'),
           )
    dic={}
    for item in patterns:
        pattern=item[0]
        area=item[1]
        match=pattern.findall(str)
        if match:
            dic[area]=int(match[-1][-1])
    return dic

def parse_time(timepoint):
    '''parses timepoint from the news text
    
    Args: 
      timepoint (string): text content of header
      
    Returns:
      datetime.datetime object in UTC timezone'''
    months={'Января':1,
            'Февраля':2,
            'Марта':3,
            'Апреля':4,
            'Мая':5,
            'Июня':6,
            'Июля':7,
            'Августа':8,
            'Сентября':9,
            'Октября':10,
            'Ноября':11,
            'Декабря':12,}
    match=re.search(r'\b(?P<day>\d+)\s(?P<month>[А-Яа-я]+)\s(?P<year>\d+)\s(?P<hour>\d+):(?P<minute>\d+)\b',timepoint,re.I)
    if match:
        point=datetime(int(match.group('year')),
                   months[match.group('month')],
                   int(match.group('day')),
                   int(match.group('hour')),
                   int(match.group('minute')),
                  ).replace(tzinfo=pytz.utc)
    try:    
        return point
    except:
        print("cannot parse timepoint from :"+timepoint)
```

## Scrapping data from coronavirus2020.kz

```python
option = webdriver.ChromeOptions()
option.add_argument("--incognito")
browser = webdriver.Chrome(executable_path = "C:/Users/showb/Documents/DS/chromedriver.exe", options=option)
browser.get('https://www.coronavirus2020.kz/ru/ofinfo');
time.sleep(2) 


#uncomment if that is first time for scrapping, and no data saved before
# for i in range(50):
#     showmore = browser.find_element_by_xpath('//*[@id="load-archive"]')
#     showmore.click()
#     time.sleep(1)    


#gathering the targeted news objects
news=browser.find_elements_by_class_name("lenta_news_title")

infected_links=[]
recovered_links=[]
for new in news: 
    if re.search(r'об эпидем.+корон',  new.text, re.I):
        infected_links.append(new.find_element_by_tag_name('a')) 
    if re.search('выздоров', new.text, re.I):
        recovered_links.append(new.find_element_by_tag_name('a'))

#extracting data from the news with number of registered cases
newcases={}
for new in infected_links: 
    new.send_keys(Keys.CONTROL + Keys.RETURN)
    browser.find_element_by_tag_name('body').send_keys(Keys.CONTROL + Keys.TAB)
    browser.switch_to.window(window_name=browser.window_handles[-1])
    wait = WebDriverWait(browser, 10)
    article = wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'article_news_body')))
    timepoint = wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'date_public_art')))
    point=parse_time(timepoint.text)
    case=parse_data(article.text)  
    browser.close()       
    browser.switch_to.window(window_name=browser.window_handles[0])    
    newcases[point]=case
    print('loaded infected statistics for',point)    

#extracting data from the news with number of registered recovering
recovered={}
for new in recovered_links: 
    new.send_keys(Keys.CONTROL + Keys.RETURN)
    browser.find_element_by_tag_name('body').send_keys(Keys.CONTROL + Keys.TAB)
    browser.switch_to.window(window_name=browser.window_handles[-1])
    wait = WebDriverWait(browser, 10)
    article = wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'article_news_body')))
    timepoint = wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'date_public_art')))
    point=parse_time(timepoint.text)
    case=parse_data(article.text)  
    browser.close()       
    browser.switch_to.window(window_name=browser.window_handles[0])
    recovered[point]=case
    print('loaded recovered statistics for',point) 
    
browser.quit()
```

## Firstly, let's fit the function to the whole portion of available data at the current time

```python
from scipy.optimize import curve_fit
target_y=np.array(infected_df.iloc[:,4:].sum().values) #first days are dropped as there is high fluctuation.
log_y=np.log(target_y) # reducing calculation dimention
x=np.arange(target_y.shape[0])
def func(x,up,s,tau):
    y=np.log(up)/( s*np.exp(-x/tau)+1)
    return y
popt, pcov = curve_fit(func, x, log_y, bounds=([10000.,0,10], [15000.,5, 30])) # the analitical guess of bounderies
print(popt)
y=func(x,*popt)
extra_x=np.arange(100)
extra_y=func(extra_x,*popt)

#natural plot
fig,ax=plt.subplots(figsize=(15,10))
ax.set_xticklabels(pd.Series(pd.date_range(start='23/3/2020', periods=20, freq='5d')).apply(lambda x: datetime.strftime(x,"%d-%m-%y")), rotation=45)
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.yaxis.set_major_locator(ticker.MultipleLocator(500))
ax.grid()
ax.set_title("Экстраполяция динамики регистрации заражений covid-19")
ax.plot(x,target_y,"go", label="официальные данные")
ax.plot(extra_x,np.exp(extra_y),"y--", label="аппроксимация функции")
ax.legend()

#plot of log
plt.figure(figsize=(15,5))
plt.title("натуральный логарифм функции")
plt.plot(x,log_y, "go")
plt.plot(extra_x,extra_y, "y--")

plt.show()
```

```python
from lmfit import Model
from sklearn.metrics import mean_squared_error

true_y=infected_df.iloc[:,4:].sum()
log_y=true_y.apply(np.log)
x=log_y.reset_index().index

x_extend=np.arange(100) #wide range of time to see the behavior of the approximated function

def func(x,up,s,tau):
    y=np.log(up)/( s*np.exp(-x/tau)+1)
    return y

model = Model(func)

fited100=model.fit(log_y, x=x, up=12174,s=0.7, tau=22.16)
y_100=fited100.eval(fited100.params, x=x_extend)

sprt=int(np.floor(0.8*x[-1]))
x_test=x[:sprt]
y_test=log_y[:sprt]
x_target=x[sprt:]
y_target=log_y[sprt:]
# fitting the data to the 80% of data using initial coefficients guess from the earlier curve_fit capture
fited80=model.fit(y_test, x=x_test, up=12174,s=0.7, tau=22.16)
# getting the forecast for the wide range of time
y_1=fited80.eval(fited80.params, x=x_extend)
mse=mean_squared_error(y_target,fited80.eval(fited80.params, x=x_target))
print("mean squared error ",np.exp(mse))

fig,ax=plt.subplots(figsize=(15,10))

ax.set_xticklabels(pd.Series(pd.date_range(start='23/3/2020', periods=20, freq='5d')).apply(lambda x: datetime.strftime(x,"%d-%m-%y")), rotation=45)
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.yaxis.set_major_locator(ticker.MultipleLocator(500))

ax.plot(x_target,np.exp(y_target),"go", label="данные валидации")
ax.plot(x_test,np.exp(y_test),"bo", label="тестовые данные")
ax.plot(x_extend,np.exp(y_1),"y--", label="Функция аппроксимированная на 80% данных")
ax.plot(x_extend,np.exp(y_100),"r--", label="Функция аппроксимированная на 100% данных")

ax.set_title("Экстраполяция динамики регистрации заражений covid-19")
ax.legend()
ax.grid()

print("parameters fo fited80 model ", fited80.params)
print("parameters fo fited100 model ", fited100.params)
plt.show()
```

## As we see some divergence of curves fitted by different portions of data, let's see the variance of such divergences using the broader testing set, somthing like Step Forward Cross Validation technique.

```python
# creating the function.
def fit_curve(gap=10, distance=7):
    '''fit the model on the portion of data defined by the two parameters.
    
      Args:
        gap: (integer): the number of points from the last testing point, after which, the difference
          between true value and calculated value will be estimated.
        distance (integer): the distance between the separator of testing 
          and evaluation data and gap
           
      Returns:
        y (array): values for fited function that are extended in time
        diver (float): the difference between actual and evaluated values after 
          the gap
        up: the parameter of model that predicts maximum number of covid-19 registrations
    '''
    try: 
        sprt=x[-1]-gap-distance
    except:
        print("distance should be between 10 and ",x[-1]-deep)
    x_test=x[:sprt]
    y_test=log_y[:sprt]
    x_target=x[sprt+gap]
    y_target=log_y[sprt+gap]
    # fitting the data to the test data using initial coefficients guess from the earlier curve_fit capture
    fited=model.fit(y_test, x=x_test, up=12174,s=0.7, tau=22.16)
    # getting the forecast for the wide range of time
    y=fited.eval(fited.params, x=x_extend)
    diver=np.exp(y_target)-np.exp(fited.eval(fited.params, x=x_target))
    diver=np.abs(diver)
    print("divergence after ",gap,"points is ",diver)
    up=fited.params.valuesdict()['up']
    return y, diver, up

fig,ax=plt.subplots(figsize=(15,10))
ax.set_xticklabels(pd.Series(pd.date_range(start='23/3/2020', periods=20, freq='5d')).apply(lambda x: datetime.strftime(x,"%d-%m-%y")), rotation=45)
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.yaxis.set_major_locator(ticker.MultipleLocator(500))

ax.plot(x,np.exp(log_y),"bo", label="данные")
ax.plot(x_extend,np.exp(y_100),"r--", label="Функция аппроксимированная на 100% данных")

divers, ups=[],[]
for i in range(10):
    y, div, up =fit_curve(distance=i)
    divers.append(div)
    ups.append(up)
    ax.plot(x_extend,np.exp(y),"--", label="Функция аппроксимированная на части данных. Точек {}".format(len(x)-10-i))


ax.set_title("Экстраполяция динамики регистрации заражений covid-19")
ax.legend()
ax.grid()

plt.show()
```
